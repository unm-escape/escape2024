### Importing Google Drive

Google CoLab and Google Drive are from the same company but they are separate "products" so they need to be linked together for the notebook to see your files. We do this by first importing the 'drive' module and mounting the drive (this will generate a bunch of permission request popups that you have to accept), and then "moving" to that folder using os.chdir.

```python
# Google Colab code to change to this directory - set your directory correctly:
from google.colab import drive
drive.mount('/content/drive')
```

```python
import os
# you will need to edit this line to make sure it is accurate for your Google Drive:
os.chdir('/content/drive/MyDrive/ESCAPE 2024/ESCAPE 2024 shared course materials/Day 1')

# check that it worked:
print(os.getcwd())

# print out the folder contents:
os.listdir()
```

### Installing python modules

Google CoLab has a good list of default modules installed, but it may sometimes be missing one that you need. In that case, you can use 'pip' to install it!

Note that because Google CoLab runs inside a virtual machine on Google servers, you will have to re-install these packages every time you close and re-open the notebook. Usually this is not a problem.

```python
# this is a very common module and is already installed by default
import numpy as np

# test that it's working
print(np.pi)
```

```python
# this one is not installed: running this cell generates an error!
# Google's built-in AI (Gemini) will offer a solution - you can click
# the "install cartopy" button and it will tell you to run "!pip install cartopy"
import cartopy
```

```python
# here it is with the install command:
!pip install cartopy
import cartopy

# check that it's working
# a simple cartopy map (suggested by Gemini):
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
ax = plt.axes(projection = ccrs.PlateCarree())
ax.coastlines()
plt.show()
```

### Coding with Gemini AI

Google's Gemini is built into CoLab and will offer suggestions in code if you write a comment or run into an error. Here are some examples of how it might be helpful:

```python
# this code has a bug, but Gemini will suggest a solution when we run it.
# click "explain error" and it will open a sidebar conversation where it will
# do its best to understand what went wrong. Pretty neat!

myVar=0
myvar="hello"
print(np.sin(myvar))
```

Here we have written a comment, and when you open the cell and press enter at the end of the line, Gemini suggests some code that it thinks we probably want to write.

Just press 'tab' to accept the suggestion. Very cool!

```python
# a short example of plotting the sin() of a variable over the range 0 to 2pi:
import numpy as np
x=np.linspace(0,2*np.pi,100)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```
