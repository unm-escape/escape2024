I have downloaded a small dataset of earthquakes in and around New Mexico from https://earthquake.usgs.gov/earthquakes/search/ with filters for M4.5+ and 1950 to now. See the file in Google Drive Folder.

Building off your exercise from Day 2, we're going to change the appearance of the plot so you can play around with changing things like font, color, grid lines, and more! Then you will have time to work on customizing your figures for your own project after this example. 

a.First, we read in this file into python using Pandas, and print out just the 'time', 'latitude', 'longitude', and 'mag' columns so we know what we are looking at. 

b. Also, make a simple plot with 'longitude' on the X axis and 'latitude' on the Y axis, using plt.scatter(). Optionally, try coloring the points or setting their size based on magnitude.

```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt

eqs = pd.read_csv('query.csv')
lat=np.array(eqs['latitude'])
lon=np.array(eqs['longitude'])
mag=np.array(eqs['mag'])
dep=np.array(eqs['depth'])

time=np.array(eqs['time'])


plt.style.use('default')
plt.scatter(lon,lat)
plt.show()
```

Let's work on exploring a few default 'themes' to get us closer to where we want to go. 'Themes' are sets of Python starting places so that you don't have to customize EVERYTHING all the time. Explore themes here: https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html

```python
plt.style.use('fivethirtyeight') #change theme to theme name of your choice!
plt.scatter(lon,lat)
plt.show()
```

We can encode information in the dot to make the figure hold more dimensions of data. For example we can encode the earthquake magnitude in size AND color

```python
plt.style.use('fivethirtyeight') #change theme to theme name of your choice!
scatter = plt.scatter(lon,lat,  c=mag, s=mag**3, cmap='Blues', alpha=0.7)
plt.show()
```

What are we looking at? Let's add a legend and axis titles!

```python
plt.style.use('fivethirtyeight') #change theme to theme name of your choice!
scatter = plt.scatter(lon,lat,  c=mag, s=mag**3, cmap='Blues', alpha=0.7)

# Label the axes
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Add color bar
plt.colorbar(scatter, label='Magnitude')

plt.show()
```

Let's customize it a bit more. Let's try to make this as basic of a plot as possible to mimize our data-ink ratio!

```python
plt.style.use('fivethirtyeight') #change theme to theme name of your choice!
scatter = plt.scatter(lon,lat,  c=mag, s=mag**3, cmap='Blues', alpha=0.7)

# Label the axes
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Add color bar
cbar=plt.colorbar(scatter, label='Magnitude')

#remove as much as possible
plt.gca().set_facecolor('none') #make plot area background transparent
plt.gcf().set_facecolor('none')  # Make the figure background transparent
for spine in plt.gca().spines.values():
    spine.set_visible(False)
cbar.outline.set_visible(False)  # Remove the outline around the color bar
cbar.ax.set_facecolor('none')    # Make the color bar's background transparent
plt.grid(False) # Remove gridlines

plt.show()
```

We can also change the font!

```python
plt.style.use('default') #change theme to theme name of your choice!

# Set the font to font of your choice
plt.rcParams['font.family'] = 'Helvetica'


scatter = plt.scatter(lon,lat,  c=mag, s=mag**3, cmap='Blues', alpha=0.7)

# Label the axes
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Add color bar
cbar=plt.colorbar(scatter, label='Magnitude')

#remove as much as possible
plt.gca().set_facecolor('none') #make plot area background transparent
plt.gcf().set_facecolor('none')  # Make the figure background transparent
for spine in plt.gca().spines.values():
    spine.set_visible(False)
cbar.outline.set_visible(False)  # Remove the outline around the color bar
cbar.ax.set_facecolor('none')    # Make the color bar's background transparent
plt.grid(False) # Remove gridlines
plt.tick_params(axis='both', which='both', length=0)  # Remove tick marks

plt.show()
```

Make sure the figure is the right size for your use case

```python
plt.style.use('default') #change theme to theme name of your choice!

# Set the font to font of your choice
plt.rcParams['font.family'] = 'Helvetica'

plt.figure(figsize=(6, 3))
scatter = plt.scatter(lon,lat,  c=mag, s=mag**3, cmap='Blues', alpha=0.7)

# Label the axes
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Add color bar
cbar=plt.colorbar(scatter, label='Magnitude')

#remove as much as possible
plt.gca().set_facecolor('none') #make plot area background transparent
plt.gcf().set_facecolor('none')  # Make the figure background transparent
for spine in plt.gca().spines.values():
    spine.set_visible(False)
cbar.outline.set_visible(False)  # Remove the outline around the color bar
cbar.ax.set_facecolor('none')    # Make the color bar's background transparent
plt.grid(False) # Remove gridlines
plt.tick_params(axis='both', which='both', length=0)  # Remove tick marks

plt.show()
```
